# Research Questions

<b> Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)

 # Performing Chi squared Test


$H_1: \ They \ are \ dependent$

$H_0: \ They \ are \ independent$

*  Compute $ \chi^2 $ 

Decide $ \alpha $

* Apply decision   
 

```python
pd.crosstab(df.Specialization,df.Gender,margins=True)
```

```python
observed=pd.crosstab(df.Specialization,df.Gender)
observed
```

```python
chi2_test_stat = scipy.stats.chi2_contingency(observed)[0]
pval = scipy.stats.chi2_contingency(observed)[1]
df = scipy.stats.chi2_contingency(observed)[2]
```

```python
confidence_level = 0.90
alpha = 1 - confidence_level
chi2_critical = scipy.stats.chi2.ppf(1 - alpha, df)
chi2_critical
```

```python
x_min = 0
x_max = 100
# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = scipy.stats.chi2.pdf(x, df)
plt.xlim(x_min, x_max)
plt.plot(x, y)
# Setting Chi2 Critical value 
chi2_critical_right = chi2_critical
# Shading the right rejection region
x1 = np.linspace(chi2_critical_right, x_max, 100)
y1 = scipy.stats.chi2.pdf(x1, df)
plt.fill_between(x1, y1, color='blue')
```

```python
if(chi2_test_stat > chi2_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```


Conclusion:

    From above research Null Hypothesis is rejected that means Both Specialization and Gender are dependent on each other.
